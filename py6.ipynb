```python
class Programmer:

    def __init__(self, name, position):
        self.name = name
        self.position = position
        self.hours = 0
        self.salary_earned = 0

        # Устанавливаем базовую ставку в зависимости от должности
        if position == "Junior":
            self.rate = 10
        elif position == "Middle":
            self.rate = 15
        elif position == "Senior":
            self.rate = 20

    def work(self, time):
        self.hours += time
        self.salary_earned += time * self.rate

    def rise(self):
        if self.position == "Junior":
            self.position = "Middle"
            self.rate = 15
        elif self.position == "Middle":
            self.position = "Senior"
            self.rate = 20
        elif self.position == "Senior":
            self.rate += 1

    def info(self):
        return f"{self.name} {self.hours}ч. {self.salary_earned}тгр."
    

programmer = Programmer('Васильев Иван', 'Junior')
programmer.work(750)
print(programmer.info())
programmer.rise()
programmer.work(500)
print(programmer.info())
programmer.rise()
programmer.work(250)
print(programmer.info())
programmer.rise()
programmer.work(250)
print(programmer.info())
```

```python
class Rectangle:

    def __init__(self, point1, point2):
        self._x1, self._y1 = point1
        self._x2, self._y2 = point2
        self._x = min(self._x1, self._x2)
        self._y = max(self._y1, self._y2)
        self._width = abs(self._x1 - self._x2)
        self._height = abs(self._y1 - self._y2)

    def perimeter(self):
        p = 2 * (self._width + self._height)
        return round(p, 2)

    def area(self):
        s = self._width * self._height
        return round(s, 2)

    def get_pos(self):
        # Округляем координаты позиции до сотых, чтобы убрать погрешность
        return (round(self._x, 2), round(self._y, 2))

    def get_size(self):
        return (round(self._width, 2), round(self._height, 2))

    def move(self, dx, dy):
        self._x += dx
        self._y += dy

    def resize(self, width, height):
        self._width = width
        self._height = height


rect = Rectangle((3.2, -4.3), (7.52, 3.14))
print(rect.get_pos(), rect.get_size())
rect.move(1.32, -5)
print(rect.get_pos(), rect.get_size())
```

```python
class NumbersError(Exception):
    pass

class EvenError(NumbersError):
    pass

class NegativeError(NumbersError):
    pass

def no_even(numbers):
    if all(x % 2 != 0 for x in numbers):
        return True
    raise EvenError("В списке не должно быть чётных чисел")

def no_negative(numbers):
    if all(x >= 0 for x in numbers):
        return True
    raise NegativeError("В списке не должно быть отрицательных чисел")

def main():
    print("Введите числа в одну строку через пробел:")
    try:
        numbers = [int(x) for x in input().split()]
        if no_negative(numbers) and no_even(numbers):
            print(f"Сумма чисел равна: {sum(numbers)}.")
    except NumbersError as e:  # Обращение к исключению как к объекту
        print(f"Произошла ошибка: {e}.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка: {e}.")
  
if __name__ == "__main__":
    main()
```

```python
import datetime as _dt
import itertools as _it
from dataclasses import dataclass, field
from enum import Enum


# =============================== КОНСТАНТЫ ===============================
DEFAULT_CARD_INFO_FIELDS = [
    "card_id",
    "user_id",
    "phone",
    "bank_name",
    "bank_bic",
    "acc_id",
    "pan",
    "payment_system",
    "currency",
    "status",
    "issue_date",
    "expiry_date",
    "balance",
    "cashback_balance",
    "user_cards"
]
DEFAULT_ACCOUNT_BALANCE = 0.00
DEFAULT_CASHBACK_BALANCE = 0.00
CARD_CURRENCY = "RUB"
DEFAULT_PAYMENT_SYSTEM = "MIR"

ACCOUNT_TYPE_CODE = "40817"     # тип счета для физлиц
ACCOUNT_BRANCH = "0000"         # отсутствие филиалов у банка
ACCOUNT_CURRENCY = "810"        # идентификатор для рублёвых операций

EMPTY_PAN = "0000000000000000"

BIN_BY_SYSTEM = {
    "MIR": "220400",
    "VISA": "400000",
    "MASTERCARD": "510000",
}

DEFAULT_CASHBACK_TRANSACTION = 0.00

TRANSACTION_HISTORY_HEADER = [
    "timestamp,type,from_card,to_card,amount,mcc,cashback,description"
]
DEPOSIT_DESCRIPTION = "{amount:.2f}₽ → карта #{card_id}"
TRANSFER_DESCRIPTION = "{amount:.2f}₽: карта #{from_card} → карта #{to_card}"
PAY_DESCRIPTION = "{amount:.2f}₽ (MCC: {mcc}) с карты #{card_id}"
BALANCE_DESCRIPTION = "Баланс: {balance:.2f}₽"


# =============================== ENUM'Ы ===============================
class CardStatus(Enum):
    ACTIVE = "Active"
    CLOSED = "Closed"
    BLOCKED = "Blocked"


CARD_STATUS = CardStatus.ACTIVE


class TransactionType(Enum):
    DEPOSIT = "deposit"
    TRANSFER = "transfer"
    PAY = "pay"
    INTEREST = "interest"


# =============================== ГЕНЕРАТОРЫ ДАННЫХ ===============================
ISSUE_DATE_START = _dt.date(2022, 1, 1)
ISSUE_DATE_GENERATOR = (
    ISSUE_DATE_START + _dt.timedelta(days=i) for i in _it.count()
)
EXPIRY_YEARS = 4

TIMESTAMP_START = _dt.datetime(2022, 1, 1, 9, 0, 0)


def timestamp_generator():
    for i in _it.count():
        base_date = TIMESTAMP_START + _dt.timedelta(days=i)
        hour = 9 + (i * 3) % 10
        minute = (i * 7) % 60
        second = (i * 11) % 60
        yield base_date.replace(hour=hour % 24, minute=minute, second=second)


TIMESTAMP_GENERATOR = timestamp_generator()


def next_timestamp_after(issue_date: _dt.date) -> _dt.datetime:
    while True:
        ts = next(TIMESTAMP_GENERATOR)
        if ts.date() > issue_date:
            return ts


# =============================== ВСПОМОГАТЕЛЬНЫЕ ФУНКЦИИ ===============================
def _luhn(pan_prefix: str) -> str:
    """Вычисляет контрольную цифру по алгоритму Луна и возвращает полный PAN."""
    digits = [int(char) for char in pan_prefix]
    sum_val = 0
    is_even = True
    
    for digit in reversed(digits):
        if is_even:
            digit *= 2
            if digit > 9:
                digit -= 9
        sum_val += digit
        is_even = not is_even
        
    check_digit = (10 - (sum_val % 10)) % 10
    return pan_prefix + str(check_digit)


def _generate_pan(payment_system: str, pan_seq: any) -> str:
    """Генерирует 16-значный PAN для заданной платежной системы."""
    bin_code = BIN_BY_SYSTEM.get(payment_system, BIN_BY_SYSTEM["MIR"])
    seq_num = next(pan_seq)
    prefix = bin_code + f"{seq_num:09d}"
    return _luhn(prefix)


def _next_account_number(bic: str, account_seq: any) -> str:
    """Генерирует номер счета по стандарту ГОСТ с использованием БИК банка."""
    seq_num = next(account_seq)
    acc_num_without_key = (
        ACCOUNT_TYPE_CODE + ACCOUNT_CURRENCY + "0" + ACCOUNT_BRANCH
        + f"{seq_num:05d}"
    )
    
    bic_slice = bic[6:8] + bic[3:6]
    full_str = bic_slice + acc_num_without_key
    
    weights = [7, 1, 3] * 8
    sum_val = sum(int(full_str[i]) * weights[i] for i in range(23))
    key = (sum_val * 7) % 10
    
    return (
        ACCOUNT_TYPE_CODE + ACCOUNT_CURRENCY + str(key) + ACCOUNT_BRANCH
        + f"{seq_num:05d}"
    )


# =============================== КЛАССЫ ===============================
@dataclass
class Transaction:
    from_card: int | None
    to_card: int | None
    amount: float
    type: str
    mcc: str | None
    description: str
    timestamp: _dt.datetime


@dataclass
class User:
    user_id: int
    last_name: str
    first_name: str
    pin: str
    phone: str
    accounts: list = field(default_factory=list)
    cards: list = field(default_factory=list)

    def change_pin(self, old_pin: str, new_pin: str):
        if self.pin == old_pin:
            self.pin = new_pin


@dataclass
class Account:
    acc_id: int
    number: str
    user_id: int
    balance: float = DEFAULT_ACCOUNT_BALANCE
    currency: str = CARD_CURRENCY
    owner: User = field(default=None, init=False)


class Card:
    def __init__(
        self,
        card_id: int,
        user: User,
        account: Account,
        payment_system: str,
        pan: str,
        issue_date: _dt.date,
        expiry_date: _dt.date,
        status: str,
        bank: "Bank"
    ):
        self.card_id = card_id
        self.user = user
        self.account = account
        self.payment_system = payment_system
        self.pan = pan
        self.issue_date = issue_date
        self.expiry_date = expiry_date
        self.status = CardStatus(status)
        self.bank = bank
        account.owner = user

    def get_card_info(self) -> str:
        user_cards_pans = [c.pan for c in self.user.cards]
        info_lines = [
            f"Карта #{self.card_id}",
            (
                f"Пользователь:  {self.user.user_id} — "
                f"{self.user.last_name} {self.user.first_name}"
            ),
            f"Телефон:       {self.user.phone}",
            f"Банк:          {self.bank.name}",
            f"БИК банка:     {self.bank.bic}",
            f"Счёт:          {self.account.number}",
            f"PAN:           {self.pan}",
            f"Плат. система: {self.payment_system}",
            f"Валюта:        {self.account.currency}",
            f"Статус:        {self.status.value}",
            f"Выпуск:        {self.issue_date.strftime('%Y-%m-%d')}",
            f"Срок:          {self.expiry_date.strftime('%Y-%m-%d')}",
            f"Баланс:        {self.account.balance:.2f}₽",
            f"Кешбэк:        {DEFAULT_CASHBACK_BALANCE:.2f}₽",
            f"Карты пользователя: {user_cards_pans}",
            "-" * 50
        ]
        return "\n".join(info_lines)

    def get_balance(self) -> str:
        return f"Баланс: {self.account.balance:.2f}₽"

    def deposit(self, amount: float):
        timestamp = next_timestamp_after(self.issue_date)
        self.account.balance += amount
        desc = DEPOSIT_DESCRIPTION.format(amount=amount, card_id=self.card_id)
        tx = Transaction(
            from_card=None,
            to_card=self.card_id,
            amount=amount,
            type=TransactionType.DEPOSIT.value,
            mcc=None,
            description=desc,
            timestamp=timestamp
        )
        self.bank.transaction_log.append(tx)

    def transfer(self, to_card: "Card", amount: float):
        latest_issue = max(self.issue_date, to_card.issue_date)
        timestamp = next_timestamp_after(latest_issue)
        self.account.balance -= amount
        to_card.account.balance += amount
        desc = TRANSFER_DESCRIPTION.format(
            amount=amount, from_card=self.card_id, to_card=to_card.card_id
        )
        tx = Transaction(
            from_card=self.card_id,
            to_card=to_card.card_id,
            amount=amount,
            type=TransactionType.TRANSFER.value,
            mcc=None,
            description=desc,
            timestamp=timestamp
        )
        self.bank.transaction_log.append(tx)

    def pay(self, amount: float, mcc: str):
        timestamp = next_timestamp_after(self.issue_date)
        self.account.balance -= amount
        desc = PAY_DESCRIPTION.format(
            amount=amount, mcc=mcc, card_id=self.card_id
        )
        tx = Transaction(
            from_card=self.card_id,
            to_card=None,
            amount=amount,
            type=TransactionType.PAY.value,
            mcc=mcc,
            description=desc,
            timestamp=timestamp
        )
        self.bank.transaction_log.append(tx)

    def close(self):
        self.status = CardStatus.CLOSED

    def get_transaction_history(self) -> list:
        sorted_txs = sorted(self.bank.transaction_log, key=lambda x: x.timestamp)
        lines = list(TRANSACTION_HISTORY_HEADER)
        for tx in sorted_txs:
            is_relevant = False
            signed_amount_str = ""
            if tx.type == TransactionType.DEPOSIT.value and tx.to_card == self.card_id:
                is_relevant = True
                signed_amount_str = f"+{tx.amount:.2f}₽"
            elif tx.type == TransactionType.PAY.value and tx.from_card == self.card_id:
                is_relevant = True
                signed_amount_str = f"-{tx.amount:.2f}₽"
            elif tx.type == TransactionType.TRANSFER.value:
                if tx.from_card == self.card_id:
                    is_relevant = True
                    signed_amount_str = f"-{tx.amount:.2f}₽"
                elif tx.to_card == self.card_id:
                    is_relevant = True
                    signed_amount_str = f"+{tx.amount:.2f}₽"

            if is_relevant:
                from_str = str(tx.from_card) if tx.from_card is not None else ""
                to_str = str(tx.to_card) if tx.to_card is not None else ""
                mcc_str = str(tx.mcc) if tx.mcc is not None else ""
                cb_str = f"{DEFAULT_CASHBACK_TRANSACTION:.2f}₽"
                ts_str = tx.timestamp.strftime("%Y-%m-%d %H:%M:%S")
                row = f"{ts_str},{tx.type},{from_str},{to_str},{signed_amount_str},{mcc_str},{cb_str},{tx.description}"
                lines.append(row)
        return lines


@dataclass
class Bank:
    name: str
    bic: str

    customers: list = field(default_factory=list, init=False)
    accounts: list = field(default_factory=list, init=False)
    cards: list = field(default_factory=list, init=False)
    transaction_log: list = field(default_factory=list)

    _user_seq: any = field(
        default_factory=lambda: _it.count(1), init=False
    )
    _account_seq: any = field(
        default_factory=lambda: _it.count(1), init=False
    )
    _card_seq: any = field(
        default_factory=lambda: _it.count(1), init=False
    )
    _pan_seq: any = field(
        default_factory=lambda: _it.count(1), init=False
    )

    def apply_for_card(
        self,
        last_name: str,
        first_name: str,
        pin: str,
        phone: str,
        payment_system: str = DEFAULT_PAYMENT_SYSTEM
    ) -> Card:
        existing_user = None
        for customer in self.customers:
            if (
                customer.phone == phone
                and customer.first_name == first_name
                and customer.last_name == last_name
            ):
                existing_user = customer
                break

        if existing_user:
            user = existing_user
        else:
            user_id = next(self._user_seq)
            user = User(
                user_id=user_id,
                last_name=last_name,
                first_name=first_name,
                pin=pin,
                phone=phone
            )
            self.customers.append(user)

        acc_id = next(self._account_seq)
        acc_number = _next_account_number(self.bic, self._account_seq)
        account = Account(
            acc_id=acc_id,
            number=acc_number,
            user_id=user.user_id,
            balance=DEFAULT_ACCOUNT_BALANCE,
            currency=CARD_CURRENCY
        )
        
        user.accounts.append(account)
        self.accounts.append(account)

        card_id = next(self._card_seq)
        pan = _generate_pan(payment_system, self._pan_seq)
        issue_date = next(ISSUE_DATE_GENERATOR)
        expiry_date = issue_date.replace(year=issue_date.year + EXPIRY_YEARS)
        
        card = Card(
            card_id=card_id,
            user=user,
            account=account,
            payment_system=payment_system,
            pan=pan,
            issue_date=issue_date,
            expiry_date=expiry_date,
            status=CARD_STATUS.value,
            bank=self
        )

        user.cards.append(card)
        self.cards.append(card)

        return card

    def get_global_history(self) -> list:
        sorted_txs = sorted(self.transaction_log, key=lambda x: x.timestamp)
        lines = list(TRANSACTION_HISTORY_HEADER)
        for tx in sorted_txs:
            from_str = str(tx.from_card) if tx.from_card is not None else ""
            to_str = str(tx.to_card) if tx.to_card is not None else ""
            amount_str = f"{tx.amount:.2f}₽"
            mcc_str = str(tx.mcc) if tx.mcc is not None else ""
            cb_str = f"{DEFAULT_CASHBACK_TRANSACTION:.2f}₽"
            ts_str = tx.timestamp.strftime("%Y-%m-%d %H:%M:%S")
            row = f"{ts_str},{tx.type},{from_str},{to_str},{amount_str},{mcc_str},{cb_str},{tx.description}"
            lines.append(row)
        return lines
    





# Заводим банк
bank = Bank("Demo Bank", "044452345")
# Заводим карты
cards = []
card_data = [
    ("Иванов", "Иван", "1234", "+79161234501", "MIR"),
    ("Петров", "Пётр", "5678", "+79161234502", "VISA"),
    ("Сидоров", "Сидор", "0000", "+79161234503", "MASTERCARD"),
    ("Кузнецов", "Кузьма", "9999", "+79161234504", "VISA"),
    ("Смирнов", "Сергей", "1111", "+79161234505", "MIR"),
    ("Смирнов", "Сергей", "1111", "+79161234505", "VISA"),
    ("Захаров", "Кирилл", "1212", "+79161234507", "MASTERCARD")
]

for last_name, first_name, pin, phone, system in card_data:
    card = bank.apply_for_card(last_name, first_name, pin, phone, system)
    cards.append(card)

# Проверяем смену PIN-Code
print("Проверка pin-code")
user = cards[0].account.owner

print("Старый PIN:", user.pin)
user.change_pin("1234", "5678")
print("Новый PIN (после правильной смены):", user.pin)

user.change_pin("0000", "9999")
print("PIN после попытки с неверным старым:", user.pin)

# Проверяем работу метода get_balance
print("\nБаланс выпущенных карт")
for i, card in enumerate(cards):
    print(card.get_balance())

# Проверяем работу метода deposit
print("\nПополняем депозиты карт")
for i, card in enumerate(cards):
    amount = 100 * (i + 1)
    card.deposit(amount)
    print(card.get_balance())

# Проверяем работу метода pay
print("\nПокупаем продукты в магазине")
print(cards[0].get_balance())
cards[0].pay(45, "5814")
print(cards[0].get_balance())

# Проверяем работу метода transfer
print("\nПереводим деньги с одной карты на другую")
print(cards[0].get_balance())
print(cards[1].get_balance())
cards[0].transfer(cards[1], 50)
print(cards[0].get_balance())
print(cards[1].get_balance())
for row in cards[1].get_transaction_history():
    print(row)

# Проверяем работу метода close
print("\nЗакрываем карту")
print(cards[0].status.value)
cards[0].close()
print(cards[0].status.value)

# Проверяем работу метода get_transaction_history
print("\nИстория транзакций банковской карты")
for row in cards[0].get_transaction_history():
    print(row)
print(cards[0].get_balance())

# Проверяем работу метода get_global_history
print("\nИстория транзакций банка")
for row in bank.get_global_history():
    print(row)
```
