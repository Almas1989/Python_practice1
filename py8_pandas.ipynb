```python
import pandas as pd


def cheque(price_list, **kwargs):
    # Сортируем позиции по названию товара (ключу kwargs) в лексикографическом порядке
    sorted_items = sorted(kwargs.items())

    products = []
    prices = []
    numbers = []
    costs = []

    for product, number in sorted_items:
        price = price_list[product]
        products.append(product)
        prices.append(price)
        numbers.append(number)
        costs.append(price * number)

    df = pd.DataFrame(
        {
            "product": products,
            "price": prices,
            "number": numbers,
            "cost": costs,
        }
    )

    return df


products = ['bread', 'milk', 'soda', 'cream']
prices = [37, 58, 99, 72]
price_list = pd.Series(prices, products)
result = cheque(price_list, soda=3, milk=2, cream=1)
print(result)
```

```python
import pandas as pd


def cheque(price_list, **kwargs):
    sorted_items = sorted(kwargs.items())

    products = []
    prices = []
    numbers = []
    costs = []

    for product, number in sorted_items:
        price = price_list[product]
        products.append(product)
        prices.append(price)
        numbers.append(number)
        costs.append(price * number)

    df = pd.DataFrame(
        {
            "product": products,
            "price": prices,
            "number": numbers,
            "cost": costs,
        }
    )

    return df


def discount(df):
    new_df = df.copy()
    new_df["cost"] = new_df.apply(
        lambda row: row["cost"] * 0.5 if row["number"] > 2 else row["cost"], axis=1
    )
    return new_df



products = ['bread', 'milk', 'soda', 'cream']
prices = [37, 58, 99, 72]
price_list = pd.Series(prices, products)
result = cheque(price_list, soda=3, milk=2, cream=1)
with_discount = discount(result)
print(result)
print(with_discount)
```
