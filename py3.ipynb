```python
from itertools import count

# Считываем три числа из ввода и преобразуем их в вещественный тип (float)
start, end, step = map(float, input().split())
print(f"Начало: {start}, Конец: {end}, Шаг: {step}")

# Запускаем бесконечный итератор с заданным началом и шагом
for x in count(start, step):
    # Округляем текущее значение, чтобы избежать ошибок точности float
    current = round(x, 2)
    
    # Проверяем условие выхода
    if current > round(end, 2):
        break
        
    # Выводим число с точностью 2 знака после запятой
    print(f"{current:.2f}")
```

```python
from itertools import chain

# Считываем пожелания трёх членов семьи и разбиваем их на списки
wishes = [input().split(', ') for _ in range(3)]
print("Пожелания членов семьи:", wishes)

# Распаковываем списки в chain, затем сортируем по алфавиту
sorted_wishes = sorted(chain(*wishes))
print("Отсортированные пожелания:", sorted_wishes)

# Нумеруем и выводим каждый продукт
for index, item in enumerate(sorted_wishes, start=1):
    print(f"{index}. {item}")
```

```python
from itertools import product

# Считываем масть, которую нужно исключить
excluded_suit = input().strip()

# Задаём номиналы и масти в нужном порядке
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'валет', 'дама', 'король', 'туз']
suits = ['пик', 'треф', 'бубен', 'червей']

# Удаляем указанную масть из списка
suits.remove(excluded_suit)

# Генерируем колоду: product автоматически переберёт каждую масть для каждого номинала
for rank, suit in product(ranks, suits):
    print(f"{rank} {suit}")
```

```python
from itertools import combinations

# Считываем количество учеников
n = int(input())

# Создаём список имён, считывая их одно за другим
names = [input() for _ in range(n)]
print("Список имён:", names)

# Функция combinations(names, 2) составляет все возможные уникальные пары
for player1, player2 in combinations(names, 2):
    print(f"{player1} - {player2}")
```

```python
from itertools import product

# Считываем общее количество долек апельсина
n = int(input())

# Выводим заголовок таблицы
print('А Б В')

# Генерируем все возможные пары долек для Ани (a) и Бори (b).
# Так как каждому нужна хотя бы 1 долька, перебираем от 1 до n-1
for a, b in product(range(1, n), repeat=2):
    # Вычисляем, сколько долек остаётся Вове
    c = n - a - b
    
    # Если Вове тоже досталась хотя бы одна долька (c > 0),
    # значит вариант успешный — выводим его
    if c > 0:
        print(f"{a} {b} {c}")
```

```python
# Считываем высоту (N) и ширину (M) прямоугольника
n = int(input())
m = int(input())

# Находим максимальное число, чтобы понять нужную ширину столбца
max_num = n * m
# Считаем, сколько символов занимает самое большое число
width = len(str(max_num))

# Создаём единственный итератор для всей последовательности чисел
linear_iter = iter(range(1, max_num + 1))

# Клонируем ссылку на этот итератор M раз и передаём в zip
for row in zip(*([linear_iter] * m)):
    # Форматируем каждое число по правому краю с вычисленной шириной 
    # и выводим через пробел (по умолчанию в print)
    print(*(f"{num:>{width}}" for num in row))
```

```python
from itertools import permutations

# Считываем количество спортсменов
n = int(input())

# Собираем имена в список и сразу сортируем по алфавиту
names = sorted(input() for _ in range(n))

# Функция permutations генерирует все возможные перестановки из списка
for arrangement in permutations(names):
    # Объединяем кортеж с именами в строку через запятую и пробел
    print(", ".join(arrangement))
```

```python
from itertools import combinations

# Словарь для перевода мастей из именительного падежа в родительный
suits_genitive = {
    "буби": "бубен",
    "пики": "пик",
    "трефы": "треф",
    "черви": "червей"
}

# Считываем обязательную масть и исключаемое достоинство
required_suit = input().strip()
excluded_rank = input().strip()

# Получаем нужное название масти для поиска
target_suit = suits_genitive[required_suit]

# Списки всех возможных достоинств и мастей (масти уже в родительном падеже)
ranks = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'валет', 'дама', 'король', 'туз']
suits = ['бубен', 'пик', 'треф', 'червей']

# Генерируем колоду без исключённого достоинства
deck = [f"{r} {s}" for r in ranks if r != excluded_rank for s in suits]

# Сортируем колоду лексикографически (как строки).
# Это автоматически поставит '10 ...' перед '2 ...', а 'бубен' перед 'пик'.
deck.sort()

# Счётчик выведенных комбинаций
count = 0

# combinations берёт карты в том порядке, в котором они лежат в deck (т.е. лексикографическом)
for trio in combinations(deck, 3):
    # Проверяем, есть ли хотя бы одна карта с нужной мастью в тройке
    if any(target_suit in card for card in trio):
        print(", ".join(trio))
        count += 1
        
        # Останавливаемся, как только вывели 10 комбинаций
        if count == 10:
            break
```

```python
def only_even(numbers):
    result = True
    for x in numbers:
        if x % 2 != 0:
           result = False
        break
    return result

print(only_even([2, 4, 6]))
print(only_even([1, 2, 3]))

# Вывод программы:
# True
# False
```
