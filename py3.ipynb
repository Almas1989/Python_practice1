```python

# Запрашиваем числа у пользователя (вводятся через пробел)
user_input = input("Введите числа через пробел: ")

# Преобразуем строку в список целых чисел с помощью генератора списка
numbers = [int(x) for x in user_input.split()]
print("Введённые числа:", numbers)

# Применяем наше выражение
odd_set = {n for n in numbers if n % 2 != 0}

# Выводим результат
print("Множество нечётных чисел:", odd_set)
```

```python
# Запрашиваем числа у пользователя (вводятся через пробел)
user_input = input("Введите числа через пробел: ")

# Преобразуем строку в список целых чисел с помощью генератора списка
numbers = [int(x) for x in user_input.split()]
print("Введённые числа:", numbers)

# Применяем наше выражение
odd_set = {n for n in numbers if n == int(n ** 0.5) ** 2}

# Выводим результат
print("Полные квадраты:", odd_set)
```

```python
numbers = [int(x) for x in input("Введите числа через пробел: ").split()]

print("Введённые числа:", numbers)
print("Максимальное произведение двух различных чисел:", max(first * second for first in numbers for second in numbers if first != second))
```

```python
# Задаём список напрямую, как настоящий объект Python, а не текст
rle = [('a', 2), ('b', 3), ('c', 1)]

# Теперь выражение отработает правильно
print(''.join(char * count for char, count in rle))
```
